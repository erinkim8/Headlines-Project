```python
import pandas as pd

# Load FinBERT-scored data
df = pd.read_csv("../data/raw/headlines_finbert.csv", parse_dates=["Date"])

# Calculate same-day and next-day returns
df["return_same_day"] = (df["CP"] - df["OP"]) / df["OP"]
df["return_next_day"] = (df["OP_next"] - df["CP"]) / df["CP"]

# Save processed version for EDA
df.to_csv("../data/processed/headlines_finbert_with_returns.csv", index=False)
df.head()
```
